```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Car_sales.csv')
df.sample(5)
```

```python
df.columns
```

**Target variable is `Sales_in_thousands`**

```python
df['Vehicle_type'].value_counts()
```

```python
df['Sales'] = df['Sales_in_thousands']*1000
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.columns
```

```python
df['__year_resale_value']
```

```python
df['year_resale_value'] = np.where(df['__year_resale_value'].isnull(),df['__year_resale_value'].mean(),df['__year_resale_value'])
```

```python
df.head()
```

```python
df.drop(['__year_resale_value'],axis=1,inplace=True)
df.dropna(axis=0,inplace=True)
df = df.reset_index(drop=True)
df.isnull().sum().sum()
```

```python
from datetime import datetime
df['Launch_year'] = pd.to_datetime(df['Latest_Launch']).dt.year
```

```python
df
```

```python
df['Launch_year'].value_counts()
```

**Sales by year**

```python
df.groupby('Launch_year')['Sales'].mean()
```

**Sales by vehicle type**

```python
df.groupby('Vehicle_type')['Sales'].count()
```

```python
sns.scatterplot(data=df,x='Sales_in_thousands',y='Price_in_thousands',hue='Launch_year',palette=plt.cm.RdYlBu)
plt.title('Sales vs Price by Launch year');
```

```python
df
```

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))

sns.histplot(data=df, x='Horsepower', ax=axes[0])
sns.histplot(data=df, x='Curb_weight', ax=axes[1])
sns.histplot(data=df, x='Width', ax=axes[2])

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.loc[:,[col for col in df.columns if df[col].dtype != 'object']].corr(),annot=True,cmap='cubehelix_r')
```
