```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('clean_weather_data.csv')
df['day'] = pd.to_datetime(df['day'], errors='coerce')
df = df.sort_values('day')
df.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['day'], df['tempMin'], label='tempMin', alpha=0.7)
plt.plot(df['day'], df['tempMax'], label='tempMax', alpha=0.7)
plt.title('Daily Temperatures')
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=df, x='tempMax', y='humidity', alpha=0.7)
plt.title('tempMax vs humidity')
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=df, x='windSpeed', y='visibility', alpha=0.7)
plt.title('windSpeed vs visibility')
plt.show()
```

```python
num_cols = ['tempMin','tempMax','cloudCover','humidity','windSpeed','visibility']
corr = df[num_cols].corr()


plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

Based on this weather correlation heatmap, here are the key insights:

â€¢ **Temperature variables are highly correlated** (0.8) - minimum and maximum temperatures move together as expected

â€¢ **Cloud cover and humidity have a moderate positive relationship** (0.58) - more clouds typically accompany higher humidity

â€¢ **Visibility decreases with humidity and cloud cover** (-0.43 and -0.26 respectively) - atmospheric moisture reduces visibility

â€¢ **Maximum temperature and humidity are negatively correlated** (-0.36) - hotter days tend to be drier

â€¢ **Wind speed shows weak relationships** with most other variables, suggesting it's largely independent of temperature and moisture conditions

â€¢ **Cloud cover and maximum temperature are negatively related** (-0.11) - clearer skies associated with higher temperatures
